```python
#here we import all the library and create a connection for creating DataBase
from sqlalchemy import create_engine, ForeignKey, Column, Integer, String , Text , Date
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
from sqlalchemy.orm import relationship
engine = create_engine("mysql+mysqlconnector://root:MosiSQL$$@localhost/Project1",echo = False)
```

```python
#at this section we declare all tables and relationships between them
class Players(Base):
    __tablename__ = 'players'
    playersId = Column(String(64), primary_key = True,autoincrement=False)
    playersName = Column(String(64))
    bornDate = Column(String(64))
    height = Column(String(64))
    foot = Column(String(64))
    nationality =Column(String(128))
    
class Teams(Base):
    __tablename__ = 'teams'

    
    legueName = Column(String(64))
    teamName = Column(String(64))
    teamId = Column(String(64), primary_key = True,autoincrement=False)
    squad = Column(String(64))
    averageAge = Column(String(64))
    foreigners = Column(String(64))
    foreignersPercent = Column(String(64))
    nationalPlayer = Column(String(64))
    teamStadium = Column(String(64))
    stadiumCapacity = Column(String(64))
    currentTransferRecord = Column(String(64))
    teamAwards = Column(String(256))
    marketValue = Column(String(64))

class League(Base):
    __tablename__ = 'league'
    id = Column(Integer, primary_key = True,autoincrement=True)
    leagueName = Column(String(64))
    season = Column(String(64))
    teamId = Column(String(64) , ForeignKey('teams.teamId'))
    teamName = Column(String(64))
    position =Column(String(64))
    matches =Column(String(64))
    total =Column(String(64))
    pts =Column(String(64))
    legueId = relationship("Teams", back_populates = "league")
Teams.league = relationship("League", order_by = League.id, back_populates = "legueId")

class PlayersLink(Base):
    __tablename__ = 'playersLink'
    id = Column(Integer, primary_key = True,autoincrement=True)
    playersId = Column(String(64),ForeignKey('players.playersId'))
    link = Column(String(512))
    linkId = relationship("Players", back_populates = "playersLink")
Players.playersLink = relationship("PlayersLink", order_by = PlayersLink.id, back_populates = "linkId")

class PlayersDetailsLink(Base):
    __tablename__ = 'playersDetailsLink'
    id = Column(Integer, primary_key = True,autoincrement=True)
    playersId = Column(String(64),ForeignKey('players.playersId'))
    link = Column(String(512))
    linkIdd = relationship("Players", back_populates = "playersDetailsLink")
Players.playersDetailsLink = relationship("PlayersDetailsLink", order_by = PlayersDetailsLink.id, back_populates = "linkIdd")





class TotalPlayers(Base):
    __tablename__ = 'totalPlayers'

    id = Column(Integer, primary_key = True,autoincrement=True)
    legueName = Column(String(64))
    teamName = Column(String(64))
    season = Column(String(64))
    teamId = Column(String(64),ForeignKey('teams.teamId'))
    playersId = Column(String(64),ForeignKey('players.playersId'))
    shirtNumber = Column(String(64))
    playersName = Column(String(64))
    position = Column(String(64))
    bornDate = Column(String(64))
    age = Column(String(64))
    nationality = Column(String(64))
    height = Column(String(64))
    currentClub = Column(String(64))
    foot = Column(String(64))
    join = Column(String(64))
    signedFrom = Column(String(64))
    contract = Column(String(64))
    marketValues = Column(String(64))
    total =relationship("Players", back_populates = "totalPlayers")
    teamID = relationship("Teams", back_populates = "totalPlayers")
Teams.totalPlayers = relationship("TotalPlayers", order_by = TotalPlayers.id, back_populates = "teamID")
Players.totalPlayers = relationship("TotalPlayers", order_by = TotalPlayers.id, back_populates = "total")
    
# ['playersName', 'playersId', 'season', 'date', 'left', 'joined', 'mv',
#        'fee']
class Transfer(Base):
    __tablename__ = 'transfer'
    id = Column(Integer, primary_key = True,autoincrement=True)
    playersName = Column(String(64))
    playersId = Column(String(64) ,ForeignKey('players.playersId'))
    season = Column(String(64))
    date = Column(String(64))
    left = Column(String(64))
    joined =Column(String(64))
    mv =Column(String(64))
    fee =Column(String(64))
    transfer = relationship("Players", back_populates = "transfer")
    
Players.transfer = relationship("Transfer", order_by = Transfer.id, back_populates = "transfer")

# ['legueName', 'teamName', 'teamId', 'squad', 'averageAge', 'foreigners',
#        'foreignersPercent', 'nationalPlayer', 'teamStadium', 'stadiumCapacity',
#        'currentTransferRecord', 'teamAwards', 'marketValue']
   
# ['playersId', 'birthDate', 'birthPlace', 'citizenship', 'height',
#        'position', 'currentValue', 'timeValue', 'season', 'competition', 'age',
#        'competitionUnicName', 'club', 'clubId', 'squad', 'appearances', 'ppg',
#        'goals', 'assists', 'ownGoal', 'subsituationsOn', 'subsituationsOff',
#        'yellowCards', 'secondYellowCards', 'redCards', 'penalyGoals',
#        'minutesPerGoal', 'minutesPlayed', 'goalsConceded', 'cleanSheets']
class PlayersFeature(Base):
    __tablename__ = 'playersFeature'

    id = Column(Integer, primary_key = True,autoincrement=True)
    playersId = Column(String(64),ForeignKey('players.playersId'))
    birthDate = Column(String(64))
    birthPlace = Column(String(64))
    citizenship = Column(String(64))
    height = Column(String(64))
    position = Column(String(64))
    currentValue = Column(String(64))
    timeValue = Column(String(64))
    season = Column(String(64))
    competition = Column(String(64))
    age = Column(String(64))
    competitionUnicName = Column(String(64))
    club = Column(String(64))
#     clubId = Column(String(64),ForeignKey('teams.teamId'))
    clubId = Column(String(64))
    squad = Column(String(64))
    appearances = Column(String(64))
    ppg = Column(String(64))
    goals = Column(String(64))
    assists = Column(String(64))
    ownGoal = Column(String(64))
    subsituationsOn = Column(String(64))
    subsituationsOff = Column(String(64))
    yellowCards = Column(String(64))
    secondYellowCards = Column(String(64))
    redCards = Column(String(64))
    penalyGoals = Column(String(64))
    minutesPerGoal = Column(String(64))
    minutesPlayed = Column(String(64))
    appearances = Column(String(64))
    goalsConceded = Column(String(64))
    cleanSheets = Column(String(64))
    featId =relationship("Players", back_populates = "playersFeature")
#     teamIdD = relationship("Teams", back_populates = "playersFeature")
# Teams.playersFeature = relationship("PlayersFeature", order_by = PlayersFeature.id, back_populates = "teamIdD")
Players.playersFeature = relationship("PlayersFeature", order_by = PlayersFeature.id, back_populates = "featId")


['playersId', 'squad', 'appearances', 'ppg', 'goals', 'assists',
       'ownGoal', 'subsituationsOn', 'subsituationsOff', 'yellowCards',
       'secondYellowCards', 'redCards', 'penalyGoals', 'minutesPerGoal',
       'minutesPlayed', 'goalsConceded', 'cleanSheets', 'position']


class TotalFeatures(Base):
    __tablename__ = 'totalFeatures'

    id = Column(Integer, primary_key = True,autoincrement=True)
    playersId = Column(String(64),ForeignKey('players.playersId'))
    squad_total = Column(String(64))
    appearances_total = Column(String(64))
    ppg_total = Column(String(64))
    goals_total = Column(String(64))
    assists_total = Column(String(64))
    ownGoal_total = Column(String(64))
    subsituationsOn_total = Column(String(64))
    subsituationsOff_total = Column(String(64))
    yellowCards_total = Column(String(64))
    secondYellowCards_total = Column(String(64))
    redCards_total = Column(String(64))
    penalyGoals_total = Column(String(64))
    minutesPerGoal_total = Column(String(64))
    minutesPlayed_total = Column(String(64))
    appearances_total = Column(String(64))
    goalsConceded_total = Column(String(64))
    cleanSheets_total = Column(String(64))
    position_total = Column(String(64))
    featureId =relationship("Players", back_populates = "totalFeatures")
Players.totalFeatures = relationship('TotalFeatures',order_by = TotalFeatures.id , back_populates = 'featureId')


Base.metadata.create_all(engine)
```
